# univariate graph

```python
fig, axes = plt.subplots(1,3, figsize =(18,5))


histogram= sns.histplot( bowling_data["wickets"],ax=axes[0] ,bins =8 , kde =True)
axes[0].set_title("histogram of economy")
histogram= sns.histplot( bowling_data["economy"],ax=axes[1] , bins =8 , kde =True)
axes[1].set_title("histogram of economy")
histogram= sns.histplot( bowling_data["avg"],ax=axes[2] , bins =8 , kde =True)
axes[2].set_title("histogram of economy")
plt.show()
```

# Bivariate Graph

```python
fig , axes=plt.subplots(1,4, figsize=(18,5))
scatter_plot1 = sns.regplot(data = bowling_data , x= "wickets",y="economy", ax = axes[0])
axes[0].set_title(" scatter plot between economy and wicktes")
scatter_plot2 = sns.regplot(data = bowling_data , x= "wickets",y="overs", ax = axes[1])
axes[1].set_title(" scatter plot between overs and wicktes")
scatter_plot2 = sns.regplot(data = bowling_data , x= "wickets",y="runs", ax = axes[2])
axes[2].set_title("scatter plot between overs and wicktes")
scatter_plot2 = sns.regplot(data = bowling_data , x= "wickets",y="dot_balls", ax = axes[3])
axes[3].set_title(" scatter plot between overs and wicktes")
plt.show()




```

```python
corr = bowling_data[
                    ["wickets",
                     "economy",
                     "overs",
                     "runs",
                     "dot_balls"]
].corr()
plt.figure(figsize =(8,6))
sns.heatmap(corr,
            annot=True,
            cmap="coolwarm",
            fmt=".2f")
plt.show()
```
