# Histogram Plot

```python
fig , axes = plt.subplots(1,4, figsize = (18,5))





histogram = sns.histplot(match_data["first_ings_score"], ax= axes[0],bins=20,kde = True)
axes[0].set_title("histogram plot")
axes[0].set_xlabel("score")
axes[0].set_ylabel("frequency")

histogram = sns.histplot(match_data["second_ings_score"], ax= axes[1],bins=20,kde = True)
axes[1].set_title("histogram plot")
axes[1].set_xlabel("score")
axes[1].set_ylabel("frequency")

histogram = sns.histplot(match_data["first_ings_wkts"], ax= axes[2],bins=10,kde = True)
axes[2].set_title("histogram plot")
axes[2].set_xlabel("wickets")
axes[2].set_ylabel("frequency")
histogram = sns.histplot(match_data["second_ings_wkts"], ax= axes[3],bins=10,kde = True)
axes[3].set_title("histogram plot")
axes[3].set_xlabel("wickets")
axes[3].set_ylabel("frequency")
plt.show()
```

```python
fig , axes = plt.subplots(1,4, figsize = (18,5))





box1 = sns.boxplot(match_data["first_ings_score"], ax= axes[0])
axes[0].set_title("box plot")


box2  = sns.boxplot(match_data["second_ings_score"], ax= axes[1])
axes[1].set_title("box plot")


box3 = sns.boxplot(match_data["first_ings_wkts"], ax= axes[2])
axes[2].set_title("box plot")

box4  = sns.boxplot(match_data["second_ings_wkts"], ax= axes[3])
axes[3].set_title("box plot")

plt.show()
```

```python
corr = match_data[
    ["first_ings_score",
     "second_ings_score",
     "first_ings_wkts",
     "second_ings_wkts"]
].corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
corr = match_data[
    ["first_ings_score",
     "second_ings_score",
     "toss_match_win"]
].corr()

plt.figure(figsize=(4,3))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
