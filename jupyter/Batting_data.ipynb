
# Univariate PLot for runs

```python



fig , axes = plt.subplots(1,3, figsize = (18,5))
histogram = sns. histplot(batting_stats_data["runs"],bins=20, ax= axes[0], kde = True)
axes[0].set_title("Distribution of Runs")
axes[0].set_xlabel("runs")
axes[0].set_ylabel("frquency")

boxplot = sns.boxplot(batting_stats_data["runs"], ax = axes[1])
axes[1].set_title("boxplot of runs")

densityplot =sns.kdeplot(batting_stats_data["runs"], fill=True, ax = axes[2])
axes[2].set_title("Density Plot of Runs")
plt.show()

```









# Bivariate Graphs

```python
fig, axes = plt.subplots(1,4, figsize = (20,5))
scatter_plot = sns.regplot(data = batting_stats_data , x = "runs", y = "strike_rate",  ax = axes[0])
axes[0].set_title("scatter plot between run and strike rate")
scatter_plot = sns.regplot(data = batting_stats_data , x = "runs", y = "fours",  ax = axes[1])
axes[1].set_title("scatter plot between run and fours")
scatter_plot = sns.regplot(data = batting_stats_data , x = "runs", y = "sixes",  ax = axes[2])
axes[2].set_title("scatter plot between run and sixes") 
scatter_plot = sns.regplot(data = batting_stats_data , x = "runs", y = "high_score",  ax = axes[3])
axes[3].set_title("scatter plot between run and high score")
```

# box plots

```python
fig, axes = plt.subplots(1, 3, figsize=(18,5))

sns.boxplot(data=batting_stats_data, x="team", y="runs", ax=axes[0])
axes[0].set_title("Runs by Team")

sns.boxplot(data=batting_stats_data, x="team", y="strike_rate", ax=axes[1])
axes[1].set_title("Strike Rate by Team")

sns.boxplot(data=batting_stats_data, x="team", y="fours", ax=axes[2])
axes[2].set_title("Fours by Team")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
